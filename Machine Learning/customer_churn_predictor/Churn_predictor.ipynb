```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/11_chrun_prediction/customer_churn.csv'
data=pd.read_csv(url)
data.drop(['customerID'],axis='columns',inplace=True)
```

```python
data[pd.to_numeric(data.TotalCharges,errors='coerce').isnull()]
```

```python
df=data[data.TotalCharges!=' ']
data.shape
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges)
```

```python
df.dtypes
```

```python
df.head()
```

```python
for column in df:
  if df[column].dtypes=='object':
    print(f'{column}: {df[column].unique()}')
```

```python
df['MultipleLines'=='No phone service']='No'
```

```python
df['MultipleLines'].replace('No phone service','No',inplace=True)
```

```python
df.replace('No internet service','No',inplace=True)
```

```python
for column in df:
  print(df[column].unique())
```

```python
df.replace('No',0,inplace=True)
df.replace('Yes',1,inplace=True)
```

```python
for column in df:
  print(df[column].unique())
```

```python
df.replace('Female',0,inplace=True)
df.replace('Male',1,inplace=True)
```

```python
for column in df:
  print(df[column].unique())
```

```python
from sklearn.preprocessing import MinMaxScaler
cols=['tenure','MonthlyCharges','TotalCharges']
scaler=MinMaxScaler()
df[cols]=scaler.fit_transform(df[cols])
```

```python
for column in df:
  print(df[column].unique())
```

```python
df1=pd.get_dummies(df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.drop('False',axis=columns,inplace=True)
```

```python
x=df1.drop(['Churn','False'],axis='columns')
y=df1['Churn']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
xtrain.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,input_shape=(26,),activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(xtrain,ytrain)
```
